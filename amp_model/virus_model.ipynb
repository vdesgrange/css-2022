```python
def simple_virus(n,t,p,infection,transmition,recovery):
    state=np.random.binomial(1, infection, n)
    k=nx.gnp_random_graph(n,p)
    A=nx.to_numpy_matrix(k)
    
    for i in range(t):
        state_new=state
        for j in range(n):
            count=0
            for w in range(n):
                if state[w]==1 and A[j,w]==1:
                    s=np.random.binomial(1, transmition)
                    if s==1:
                        state_new[j]=1
                        count+=1
            if count==0:
                s=np.random.binomial(1, recovery)
                if s==1:
                    state_new[j]=0
        state=state_new
        colors=["red" if j==1 else "blue"  for j in state]
        nx.draw(k, node_color=colors)
        return(state)
```

```python
simple_virus(20,10,0.3,0.4,0.6,0.8)
```
